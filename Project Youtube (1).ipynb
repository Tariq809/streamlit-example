```python
api_service_name = "youtube"
api_version = "v3"
```

```python
youtube = build(api_service_name, api_version, developerKey=api_key)
request = youtube.channels().list(part='snippet,contentDetails,statistics', id=channel_id)
response = request.execute();
```

```python
print(response)
```

```python
import json
print(json.dumps(response, indent=4))
```

```python
playlist_id = response['items'][0]['contentDetails']['relatedPlaylists']['uploads']
```

```python
playlistReq = youtube.playlistItems().list(part='contentDetails,snippet,status', playlistId=playlist_id)
```

```python
playlistRes = playlistReq.execute();
```

```python
print(playlistRes)
```

```python
print(json.dumps(playlistRes, indent=2))
```

```python
video_id = playlistRes['items'][0]['contentDetails']['videoId']
videoReq = youtube.videos().list(part="snippet,contentDetails,statistics",id=video_id)
videoRes = videoReq.execute();
print(json.dumps(videoRes, indent=2))
```

```python
commentReq = youtube.commentThreads().list(part="snippet", videoId=video_id);
commentRes = commentReq.execute();
print(json.dumps(commentRes, indent=2))
```
